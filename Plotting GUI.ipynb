```python
from tkinter import *
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg) 
import numpy as np
import pandas as pd
global var
root=tk.Tk()
root.title("FUNCTION PLOTTER")
root.geometry("600x700")
var=StringVar(root,"1")
#PLOT FUNCTION
def plot():
    global var
    #IF USER INPUT IS FUNCTIONAL PLOT
    if var.get()=="Functional Plot":
        label_1 = tk.Label(root, text="function name:",width=20,font=("bold", 10))
        label_1.place(relx = 0, rely = 0.32)
        list1=['sin','exponential','polynomial','logarithm']
        #OPTION MENU TO CHOOSE THE FUNCTIONAL PLOT TYPE
        global var1
        var1=StringVar()
        droplist=OptionMenu(root,var1,*list1)
        var1.set("Select Function")
        droplist.config(width=15)
        droplist.place(relx = 0.25, rely = 0.32)
        #PLOTTER CLASS WHICH CONTAINS DIFFERENT PLOTTING METHODS
        class plotter():
            def sin():
                root3=Tk()
                root3.title("Sine Plot")
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                t = np.arange(int(entry_2.get()), int(entry_3.get()), .01)
                fig.add_subplot(111).plot(t, np.sin(s1.get()*t))
                canvas = FigureCanvasTkAgg(fig, master=root3)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH,expand=1) 
            def exponential():
                root3=Tk()
                root3.title("Exponential Plot")
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                t = np.arange(int(entry_2.get()), int(entry_3.get()), .01)
                fig.add_subplot(111).plot(t,np.exp(s2.get()*t))
                canvas = FigureCanvasTkAgg(fig, master=root3)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1) 
            def polynomial():
                root3=Tk()
                root3.title("Polynomial-degree 2 Plot")
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                t = np.arange(int(entry_2.get()), int(entry_3.get()), .01)
                fig.add_subplot(111).plot(t,((int(entry_x2.get())*(t**2))+(int(entry_x.get())*t)+int(entry_c.get())))
                canvas = FigureCanvasTkAgg(fig, master=root3)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)  
            def logarithm():
                root3=Tk()
                root3.title("Logarithm Plot")
                if float(entry_2.get())>0 and float(entry_3.get())>0 and entry_b.get()=='e':
                    fig = plt.Figure(figsize=(2, 4), dpi=100)
                    t = np.arange(float(entry_2.get()), float(entry_3.get()),.01)
                    fig.add_subplot(111).plot(t,(np.log(t)))
                    canvas = FigureCanvasTkAgg(fig, master=root3)  # A tk.DrawingArea.
                    canvas.draw()
                    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)                 
                elif float(entry_2.get())>0 and float(entry_3.get())>0:
                    fig = plt.Figure(figsize=(2, 4), dpi=100)
                    t = np.arange(float(entry_2.get()), float(entry_3.get()),.01)
                    fig.add_subplot(111).plot(t,(np.log(t)/np.log(int(entry_b.get()))))
                    canvas = FigureCanvasTkAgg(fig, master=root3)  # A tk.DrawingArea.
                    canvas.draw()
                    canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=1)  
                else:
                    root3.geometry("400x100")
                    label_error = Label(root3, text="ERROR",width=20,font=("bold", 10))
                    label_error.place(x=150,y=10)
        def parameter():
            #ENTER THE RANGE OF X-AXIS
            label_2 = Label(root, text="start:",width=20,font=("bold", 10))
            label_2.place(relx=0,rely=0.42)
            global entry_2
            entry_2=Entry(root)
            entry_2.get()
            entry_2.place(relx=0.25,rely=0.42)

            label_3 = Label(root, text="End:",width=20,font=("bold", 10))
            label_3.place(relx=0,rely=0.52)
            global entry_3
            entry_3=Entry(root)
            entry_3.get()
            entry_3.place(relx=0.25,rely=0.52)
            #ENTER THE PARAMETERS OF FUNCTIONS DEPENDING UPON THE INPUT
            if var1.get()=='sin':
                v1 = DoubleVar() 
                global s1
                s1 = Scale( root, variable = v1,  
                   from_ = 1, to = 100,  
                   orient = HORIZONTAL) 
                l3 = Label(root, text = "Scaler") 
                s1.place(relx=0.25,rely=0.62)
                l3.place(relx=0.25,rely=0.69)
                #BUTTON TO GO TO PLOT PROCEDURE
                but_login=Button(root, text='plot',bg='brown',fg='white',command=plotter.sin).place(relx=0.25,rely=0.79)
            elif var1.get()=='exponential':
                v2 = DoubleVar() 
                global s2
                s2 = Scale( root, variable = v2,  
                   from_ = -50, to = 50,  
                   orient = HORIZONTAL) 
                l4 = Label(root, text = "power parameter") 
                s2.place(relx=0.25,rely=0.62)  
                l4.place(relx=0.25,rely=0.69)
                #BUTTON TO GO TO PLOT PROCEDURE
                but_login3=Button(root, text='plot',bg='brown',fg='white',command=plotter.exponential).place(relx=0.25,rely=0.79)                
            elif var1.get()=='polynomial':
                label_x2 = Label(root, text="X2:",width=20,font=("bold", 10))
                label_x2.place(relx=0,rely=0.62)
                global entry_x2
                entry_x2=Entry(root)
                entry_x2.place(relx=0.25,rely=0.62) 
                
                label_x = Label(root, text="X:",width=20,font=("bold", 10))
                label_x.place(relx=0,rely=0.72)
                global entry_x
                entry_x=Entry(root)
                entry_x.place(relx=0.25,rely=0.72)
                
                label_c = Label(root, text="C:",width=20,font=("bold", 10))
                label_c.place(relx=0,rely=0.82)
                global entry_c
                entry_c=Entry(root)
                entry_c.place(relx=0.25,rely=0.82)
                #BUTTON TO GO TO PLOT PROCEDURE
                but_login1=Button(root, text='plot',bg='brown',fg='white',command=plotter.polynomial)
                but_login1.place(relx=0.25,rely=0.90)
            elif var1.get()=='logarithm':
                label_b = Label(root, text="Base:",width=20,font=("bold", 10))
                label_b.place(relx=0,rely=0.72)
                global entry_b
                entry_b=Entry(root)
                entry_b.place(relx=0.25,rely=0.72)
                #BUTTON TO GO TO PLOT PROCEDURE
                but_login4=Button(root, text='plot',bg='brown',fg='white',command=plotter.logarithm)
                but_login4.place(relx=0.25,rely=0.82) 
        #BUTTON TO CHOOSE PARAMETERS FOR A PARTICULAR FUNCTIONAL PLOT
        but_login2=Button(root, text='Next',bg='brown',fg='white',command=parameter).place(relx = 0.60, rely = 0.32)
    #IF USER INPUT IS STATISTICAL PLOT
    else:
        T = Text(root,height = 5, width =100, bg="white")
        T.place(x=0,y=280)
        l = Label(root,text = "Name of the File")
        l.place(x=250,y=230)
        #GRAPHER CLASS WHICH CONTAINS DIFFERENT PLOTTING METHODS
        class grapher():
            def hist():
                rootj=Tk()
                rootj.geometry('400x800')
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                fig.add_subplot(111).hist(df[varp.get()])
                canvas = FigureCanvasTkAgg(fig, master=rootj)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH,expand=1)    
                rootj.mainloop()
            def pie():
                rootj=Tk()
                rootj.geometry('400x800')
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                lis=list(df[vark.get()])
                fig.add_subplot(111).pie(df[varp1.get()],labels=lis,autopct='%1.1f%%')
                canvas = FigureCanvasTkAgg(fig, master=rootj)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH,expand=1)    
                rootj.mainloop()
            def X_Yplot():
                rootj=Tk()
                rootj.geometry('400x800')
                fig = plt.Figure(figsize=(2, 4), dpi=100)
                lis=list(df[vark2.get()])
                fig.add_subplot(111).bar(df[varp2.get()],lis)
                canvas = FigureCanvasTkAgg(fig, master=rootj)  # A tk.DrawingArea.
                canvas.draw()
                canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH,expand=1)    
                rootj.mainloop()                
        def plotterb():
            #ONLY ONE COLUMN FOR HISTOGRAM PLOT
            if Type.get()=='Histogram Plot':
                x=list(df.columns)
                label_1=tk.Label(root,text="column",width=20,font=("bold", 10))
                label_1.place(relx=0,rely=0.82)
                global varp
                varp=StringVar()
                #OPTION MENU FOR CHOOSING THE COLUMN
                droplist=OptionMenu(root,varp,*x)
                varp.set("select column")
                droplist.config(width=15)
                droplist.place(relx=0.20,rely=0.82)
                b2=Button(root,text="plot",bg='brown',fg='white',command=grapher.hist)
                b2.place(relx=0.20,rely=0.9)
            #CHOOSE LABEL COLUMN AND SECTOR COLUMN FOR THE PIE PLOT
            elif Type.get()=='Pie Plot':
                x=list(df.columns)
                label_1=tk.Label(root,text="column",width=20,font=("bold", 10))
                label_1.place(relx=0,rely=0.82)
                global varp1
                varp1=StringVar()
                #OPTION MENU FOR CHOOSING THE SECTOR DATA
                droplist=OptionMenu(root,varp1,*x)
                varp1.set("select column")
                droplist.config(width=15)
                droplist.place(relx=0.20,rely=0.82)
                global vark
                vark=StringVar()
                label_2=tk.Label(root,text="Label",width=20,font=("bold", 10))
                label_2.place(relx=0.50,rely=0.82)
                varp=StringVar()
                #OPTION MENU TO CHOOSE THE COLUMN TO LABEL
                droplist=OptionMenu(root,vark,*x)
                vark.set("select Label")
                droplist.config(width=15)
                droplist.place(relx=0.70,rely=0.82)                
                b2=Button(root,text="plot",bg='brown',fg='white',command=grapher.pie)
                b2.place(relx=0.20,rely=0.9)
            #CHOOSE X-AXIS AND Y-AXIS COLUMNS FOR THE PLOT
            elif Type.get()=='X-Y Plot':
                x=list(df.columns)
                label_1=tk.Label(root,text="X-axis",width=20,font=("bold", 10))
                label_1.place(relx=0,rely=0.82)
                global varp2
                varp2=StringVar()
                #OPTIONMENU TO CHOOSE X-AXIS
                droplist=OptionMenu(root,varp2,*x)
                varp2.set("X-axis")
                droplist.config(width=15)
                droplist.place(relx=0.20,rely=0.82)
                global vark2
                vark2=StringVar()
                label_2=tk.Label(root,text="Y-axis",width=20,font=("bold", 10))
                label_2.place(relx=0.50,rely=0.82)
                #OPTIONMENU TO CHOOSE Y-AXIS
                droplist=OptionMenu(root,vark2,*x)
                vark2.set("Y-axis")
                droplist.config(width=15)
                droplist.place(relx=0.70,rely=0.82)                
                b2=Button(root,text="plot",bg='brown',fg='white',command=grapher.X_Yplot)
                b2.place(relx=0.20,rely=0.9)                
        def choosetype():
            global df
            df=pd.read_csv(T.get("1.0",'end-1c').replace("'\'",'/'))
            label_type=Label(root, text = "Type Of Plot",
                      font = ("Times New Roman", 10)).place(relx=0,rely=0.65)
            global Type
            Type = tk.StringVar()
            #COMBOBOX TO LIST THE AVAILABLE PLOTS
            Typechoosen = ttk.Combobox(root, width = 27, textvariable = Type)
            Typechoosen['values'] = ('Histogram Plot','Pie Plot','X-Y Plot')
            Typechoosen.place(relx=0.20,rely=0.65)
            #BUTTON TO PROCEED TO FUNCTION WHERE PLOT PARAMETERS ARE GIVEN
            b2=Button(root,text='Next',command=plotterb)
            b2.place(relx=0.20,rely=0.75)
        #BUTTON TO PROCEED TO CHOOSE THE APPROPRIATE PLOT
        b1=Button(root,text = "Next",command=choosetype)
        b1.place(relx=0.5,y=400)
#RADIO BUTTONS TO CHOOSE BETWEEN PLOT TYPE
label1=tk.Label(root,text="Choose Type Of Plot:",fg="#ff8811")
label1.place(relx = 0, rely = 0)
button1=tk.Radiobutton(root,text="Statistical Plot",variable=var,value="Statistical Plot")
button1.place(relx = 0.07, rely = 0.07)
button2=tk.Radiobutton(root,text="Functional Plot",variable=var,value="Functional Plot")
button2.place(relx = 0.07, rely = 0.15)
#BUTTON TO PROCEED TO PLOT COMMAND
b5=tk.Button(root,text="Proceed",command=plot)
b5.place(relx = 0.2, rely = 0.25)
root.mainloop()
```
